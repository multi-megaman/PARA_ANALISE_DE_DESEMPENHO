```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML

# Load the CSV file into a DataFrame
file_path_1 = '50p_2000r_20x.csv'
df1 = pd.read_csv(file_path_1)
# Extract statistics of each column
statistics1 = df1.describe()

# Load the CSV file into a DataFrame
file_path_2 = '5000p_2000r_20x.csv'
df2 = pd.read_csv(file_path_2)
# Extract statistics of each column
statistics2 = df2.describe()

# Display both tables side by side with titles
display(HTML(f"""
<table>
    <tr>
        <th>Statistics for 50p_2000r_20x.csv</th>
        <th>Statistics for 5000p_2000r_20x.csv</th>
    </tr>
    <tr>
        <td>{statistics1.to_html()}</td>
        <td>{statistics2.to_html()}</td>
    </tr>
</table>
"""))
```
